```python
# Crafted 1-12-21 : Jeffrey T. Scott
# Initial imports
import pandas as pd 
import numpy as np
import datetime as dt
from pathlib import Path
import seaborn as sns
%matplotlib inline
# inital Path of Data
whale_path=Path('whale_returns.csv')
sp500_path=Path('sp500_history.csv')
algo_path=Path('algo_returns.csv')
```

```python
## THIS BOX READS AND FORMATS WHALE RETURNS ONLY!!!!
# Reading whale returns
whale_df = pd.read_csv(whale_path
                        , index_col = 'Date'
                        , infer_datetime_format = True
                        , parse_dates = True
                        ).sort_index()
# Count nulls
whale_df.isnull().sum()
# Drop nulls
whale_df = whale_df.dropna()
# Check Data:describe();shape();keys();index();info();dtypes()
whale_df.info() # isnull() works I prefer info()
```

```python
## THIS BOX READS AND FORMATS ALGO RETURNS ONLY!!!!
# Reading algorithmic returns
algo_df = pd.read_csv(algo_path
                    , index_col = 'Date'
                    , infer_datetime_format = True
                    , parse_dates = True
                    ).sort_index()
# Count nulls
algo_df.isnull().sum()
# Drop nulls
algo_cleandf = algo_df.dropna()
# Check Data/
algo_cleandf.info()              
```

```python
## THIS BOX READS AND FORMATS S&P 500 RETURNS ONLY!!!!
# Reading S&P 500 Closing Prices
sp500df = pd.read_csv(sp500_path
                    , index_col = 'Date'
                    , infer_datetime_format = True
                    , parse_dates = True
                    ).sort_index()
# Count Nulls
sp500df.isnull().sum().sum()
# Fix Data Types
sp500df["Close"] = sp500df["Close"].str.replace("$", "").astype("float")
# Calculate Daily Returns
sp500dfrtns = sp500df.pct_change()
# Drop nulls & Rename `Close` Column to be specific to this portfolio.
sp500cleaned = sp500dfrtns.dropna().rename(columns = {"Close" : "S&P500 Returns"})
sp500cleaned.info()
```

```python
# Join Whale Returns, Algorithmic Returns, and the S&P 500 Returns into a single DataFrame with columns for each portfolio's returns.
combined_rtns = pd.concat([sp500cleaned,algo_cleandf,whale_cleandf], axis = 'columns', join = 'inner')
combined_rtns.info()
```

```python
# Plot daily returns of all portfolios
combined_rtns.plot(
            kind = 'line'
            , figsize = (20,20)
            , grid = True
            , subplots = True
            , fontsize = 12
            , ylim = [-.06,.06]
            , rot = 0
            , lw = 2
            )    
```

```python
# Calculate cumulative returns of all portfolios
cum_rets = (1 + combined_rtns).cumprod()
cum_rets.plot(
        kind = 'line'
        , figsize = (25,10)
        , grid = True
        , title = "Portfolio Returns"
        , ylabel = "Total Rtns"
        , xlabel = 'Dates'
        , rot = 0
        , fontsize = 15
        , color = ['red','g','y','b','orange','pink','black']
    )
```

```python
# Box plot to visually show risk
cum_rets.plot(
    kind='box'
    , figsize=(25,10)
    , grid=True 
    , title="Portfolio Risk : based on cumualtive returns"
    #, ylim=[-.2,.2]
    , ylabel="Risk Percentile"
    , rot=0
    , whis = [0,100]
    #, fontsize=15
) 
```

```python
# Calculate the daily standard deviations of all portfolios
daily_std=combined_rtns.std().sort_values()
daily_std
```

```python
daily_stdchart= pd.concat([daily_std],axis='columns',join='outer')
bins = [0, 0.008553, 0.008555, 1]
group_names = ["Lower", "Benchmark", "Higher"]
daily_stdchart["Risk Level vs S&P500"] = pd.cut(daily_stdchart[0], bins, labels=group_names)
daily_stdchart.columns = ['STD','Risk Level']
# Calculate  the daily standard deviation of S&P 500
daily_std['S&P500 Returns'].round(6)
# Determine which portfolios are riskier than the S&P 500
print(f'The following Portfolios are riskier than the SP500: \n\n '
      f'{daily_stdchart.iloc[5::]}')
```

```python
# Calculate the annualized standard deviation (252 trading days)
annual_std=daily_std*np.sqrt(252)
annual_std.sort_values(inplace=True)
annual_std
```

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
roll_twenty_one=combined_rtns.rolling(window=21).std().dropna()
# Plot the rolling standard deviation
roll_twenty_one.plot(
              kind='line'
            , figsize=(30,20)
            , grid=True
            #, subplots=True
            , fontsize=12
            , ylim=[0,.035]
            , rot=0
            , lw=1.5
            , color=['r','b','b','g','g','g','g']
        )
```

```python
# Calculate the correlation
correlation=combined_rtns.corr()
# Display the correlation matrix
print(correlation.sum(axis=0).sort_values())

sns.heatmap(correlation, vmin=0, vmax=1,annot=True,)
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half-life.

```python
# Use `ewm` to calculate the rolling window
dfewa = df1.ewm(halflife = '21', times = (df1.index)).mean()
dfewa.columns = ['Algo 1','BERKSHIRE HATHAWAY INC','SOROS FUND MANAGEMENT LLC']
dfewa.plot(subplots = False
            , figsize = (30,20)
            , grid = True
            , ylim = (-1,5))
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
sharpe1_ratio = ((combined_rtns.mean())*252) / (combined_rtns.std()*np.sqrt(252))
sharpe1_ratio.sort_values(inplace = True)
sharpe1_ratio
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe1_ratio.plot(
                kind = 'bar'
                , rot = 360
                , figsize = (25,10)
                , color = ['r','r','y','y','y','y','g'])
```

```python
# Reading data from 1st stock
googdf=pd.read_csv('goog_historical.csv'
                        , index_col = 'Trade DATE'
                        , infer_datetime_format = True
                        , parse_dates = True
                    ).sort_index()
googdf=googdf['NOCP']
googdf=googdf.rename('Google')
```

```python
# Reading data from 2nd stock
aapldf=pd.read_csv('aapl_historical.csv',index_col='Trade DATE', infer_datetime_format=True, parse_dates=True)
aapldf=aapldf['NOCP']
aapldf=aapldf.rename('Apple')
```

```python
# Reading data from 3rd stock
costdf=pd.read_csv('cost_historical.csv',index_col='Trade DATE', infer_datetime_format=True, parse_dates=True)
costdf=costdf['NOCP']
costdf=costdf.rename('Costco')
```

```python
# Combine all stocks in a single DataFrame
combineddf=pd.concat([costdf,googdf,aapldf],axis='columns',join='inner')
```

```python
# Rename Date index
myportcleaned=combineddf.sort_index().pct_change().dropna()
myportcleaned=myportcleaned.rename_axis('Date')
myportcleaned.tail()
```

```python
# Set weights
weights = [1/3, 1/3, 1/3]
# Calculate portfolio return
my_portfolio_returns=myportcleaned.dot(weights)
my_portfolio_returns.head()
```
